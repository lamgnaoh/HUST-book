```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import os.path
from os import path
from keras.models import Sequential
from keras.layers import Dense
from keras.layers import LSTM
from keras.layers import Dropout
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
pip install -U scikit-learn scipy matplotlib
```

```python
training_set = df_train.iloc[:, 6:7].values#lấy ra tập gồm các giá trị thuộc cột Mean trong df_train
sc = MinMaxScaler(feature_range = (0, 1))
training_set_scaled = sc.fit_transform(training_set)# chuẩn hóa tập training_set thành training_set_scaled các giá trị
#trong khoảng 0-1
```

```python
training_set
```

```python
training_set_scaled
```

```python
Tạo dữ liệu train, X = 60 time steps, Y =  1 time step, có nghĩa lấy dữ liệu của 60 ngày trước để dự đoán cho ngày tiếp theo 
hay có thể hiểu là lấy 60 ngày trước làm đầu vào, và đầu ra của mô hình sẽ là ngày tiếp theo
```

```python
X_train = []
y_train = []
no_of_sample = len(training_set)

for i in range(60, no_of_sample):
    X_train.append(training_set_scaled[i-60:i, 0])#dữ liệu của 60 ngày trước
    y_train.append(training_set_scaled[i, 0])# dữ liệu của ngày tiếp theo

X_train, y_train = np.array(X_train), np.array(y_train)

X_train = np.reshape(X_train, (X_train.shape[0], X_train.shape[1], 1))
```

```python
X_train.shape[1] # số cột
```

```python
X_train.shape[0] # số hàng
```

```python
X_train
```

```python
y_train
```

```python
# Xây dựng model LSTM
regressor = Sequential()
regressor.add(LSTM(units = 50, return_sequences = True, input_shape = (X_train.shape[1], 1)))
regressor.add(Dropout(0.2))
regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(0.2))
regressor.add(LSTM(units = 50, return_sequences = True))
regressor.add(Dropout(0.2))
regressor.add(LSTM(units = 50))
regressor.add(Dropout(0.2))
regressor.add(Dense(units = 1))
regressor.compile(optimizer = 'adam', loss = 'mean_squared_error')
```

```python
# Nếu tồn tại model thì load lại nếu không thì thực hiện train 
if path.exists("mymodel.h5"):
    regressor.load_weights("mymodel.h5")
else:
    regressor.fit(X_train, y_train, epochs = 100, batch_size = 32)
    regressor.save("mymodel.h5")
```

```python
real_stock_price = df_test.iloc[:, 6:7].values
```

```python
real_stock_price
```

```python
dataset_total = pd.concat((df_train['Mean'], df_test['Mean']), axis = 0)
inputs = dataset_total[len(dataset_total) - len(df_test) - 60:].values# len(df_test) = 252 ,len(dataset_total) = 5579
inputs = inputs.reshape(-1,1)
inputs = sc.transform(inputs)

X_test = []
no_of_sample = len(inputs)

for i in range(60, no_of_sample):
    X_test.append(inputs[i-60:i, 0])

X_test = np.array(X_test)
X_test = np.reshape(X_test, (X_test.shape[0], X_test.shape[1], 1))
predicted_stock_price = regressor.predict(X_test)
predicted_stock_price = sc.inverse_transform(predicted_stock_price)
```
